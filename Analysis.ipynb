```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df = df.drop("Cabin", axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
sex_map = {"male":1, "female":0}
df['Sex'] = df['Sex'].map(sex_map)
df
```

```python
mean_age = df["Age"].mean()
mean_fare = df["Fare"].mean()
df['Age'].fillna(mean_age, inplace = True)
df['Fare'].fillna(mean_fare, inplace = True)
df
```

```python
for x in [1,2,3]:    ## for 3 classes
    df.Age[df.Pclass == x].plot(kind="kde")
plt.title("Age wrt Pclass")
plt.legend(("1st","2nd","3rd"))
plt.show()
```

```python
sns.factorplot(x="Age", y="Embarked",
               hue="Sex", row="Pclass",
               data=df,
               orient="h", size=2, aspect=3.5,kind = 'violin', split = True)
```

## Decision Tree

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import recall_score
```

```python
x = df[["Pclass","Sex","Age","SibSp","Parch","Fare"]]
y = df["Embarked"]
```

```python
sns.heatmap(x.corr(), cmap='magma', linecolor='white', linewidth=1, annot = True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.2, random_state = 6)

from sklearn import tree
Decision_tree = tree.DecisionTreeClassifier(random_state = 2)
```

```python
Decision_tree.fit(x_train,y_train)
```

```python
predictions = Decision_tree.predict(x_test)
acc_1 = accuracy_score(y_test, predictions)
print("Accuracy")
print(accuracy_score(y_test, predictions),'\n')
print('Confusion matrix')
print(confusion_matrix(y_test, predictions),'\n')
print('Classification Report')
print(classification_report(y_test, predictions))
print('Recall')
print(recall_score(y_test, predictions,average = 'macro'))
```

## K-Nearest Neighbor

```python
x = df[["Pclass","Age","SibSp","Parch","Fare"]]
y = df["Embarked"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.2, random_state = 6)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 13)
classifier.fit(x_train, y_train)
```

```python
prediction_2 = classifier.predict(x_test) 
```

```python
prediction_2
```

```python
print("Accuracy")
acc_2 = accuracy_score(y_test, prediction_2)
print(accuracy_score(y_test, prediction_2),'\n')
print("Confusion Matrix")
print(confusion_matrix(y_test, prediction_2),'\n')
print("Classification Report")
print(classification_report(y_test, prediction_2))
```

## Random Forest

```python
x = df[["Pclass","Age","SibSp","Parch","Fare"]]
y = df["Embarked"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.2, random_state = 6)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 17, random_state = 1)
classifier.fit(x_train, y_train)
```

```python
prediction_3 = classifier.predict(x_test)
```

```python
print("Accuracy")
acc_3 = accuracy_score(y_test, prediction_3)
print(accuracy_score(y_test, prediction_3),'\n')
print("Confusion Matrix")
print(confusion_matrix(y_test, prediction_3),'\n')
print("Classification Report")
print(classification_report(y_test, prediction_3))
```

## Comparing the Algorithms

```python
results = pd.DataFrame(columns = ['Decision Tree', 'K-NN','Random Forest'], index = ['Accuracy'])
results['Decision Tree'] = acc_1
results['K-NN'] = acc_2
results['Random Forest'] = acc_3
results
```
